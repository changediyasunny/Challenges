```python
"""
Given two string s and t and a positive integer K, determine whether we can swap
any two characters in string s at most K times so that s is equal to t.

Example 1

s = "converse"
t = "conserve"
K = 1
>> return true.
because we can swap v and s in string s and we swap only once
which is not greater than K times.

Example 2

s = "aabbccdd"
t = "ddbbccaa"
K = 3

>> return true.
because string s can be converted to string t in two swaps which is not
greater than K. But in example 2, if string s and string t are still the same but K
becomes 1, then we should return false.

The length of string s and t are guaranteed to be the same.

"""
def kSimilarity(strs1, strs2, k):
    def neighbors(strs):
        print(strs)
        for i, c in enumerate(strs):
            if c != strs2[i]:
                break
        data = []
        temp_strs = list(strs)
        for j in range(i+1, len(strs)):
            if strs[j] == strs2[i]:
                temp_strs[i], temp_strs[j] = temp_strs[j], temp_strs[i]
                data.append("".join(temp_strs))
                temp_strs[j], temp_strs[i] = temp_strs[i], temp_strs[j]
        print(">> neighbours: %s" %data)
        return data

    queue = deque([strs1])
    seen = {strs1: 0}
    while queue:
        strs = queue.popleft()
        if strs == strs2:
            if seen[strs] > k:
                return False
            return True
                
        for tmp_strs in neighbors(strs):
            if tmp_strs not in seen:
                seen[tmp_strs] = seen[strs] + 1
                queue.append(tmp_strs)

strs1 = "aabbccdd"
strs2 = "ddbbccaa"
K = 3

print(kSimilarity(strs1, strs2, K))
```

```python
"""
Consider a binary tree, where an arbitary node has 2 parents i.e two nodes in the tree have the same child.

Identify the defective node with 2 parents.
Correct such a node and restore the binary tree properties to that node.

"""

# Binary Tree
def find_defect(root):
    stack = [root]
    visited = {root.val}
    while stack:
        curr = stack.pop()
        if curr.left:
            if curr.left.val in visited:
                print("Left defected node: %s" %curr.left.val)
                curr.left = None
            else:
                visited.add(curr.left.val)
                stack.append(curr.left)
        if curr.right:
            if curr.right.val in visited:
                print("Right defected node: %s" %curr.right.val)
                curr.right = None
            else:
                visited.add(curr.right.val)
                stack.append(curr.right)
    return root
```

```python
""" 730. Count Different Palindromic Subsequences """
def palindrome_subsequence(strs):
    
    n = len(strs)
    table = [[0 for _ in range(n)] for _ in range(n)]
    
    # single char palindrome
    for i in range(n):
        table[i][i] = 1
    pp.pprint(table)
    for k in range(1, n):
        for i in range(n-k):
            j = i + k
            if strs[i] == strs[j]:
                # chars are equal
                low = i + 1
                high = j - 1
                while low <= high and strs[low] != strs[j]:
                    low += 1
                while low<=high and strs[high] != strs[j]:
                    high -= 1
                
                if low > high:
                    # 'aba'
                    table[i][j] = table[i + 1][j - 1] * 2 + 2;
                elif low == high:
                    # 'aaa'
                    table[i][j] = table[i + 1][j - 1] * 2 + 1;
                else:
                    # 'aacaa'
                    table[i][j] = table[i + 1][j - 1] * 2 - table[low + 1][high - 1];
            else:
                # chars are not equal
                table[i][j] = table[i][j - 1] + table[i + 1][j] - table[i + 1][j - 1];
    return table[0][n-1]


palindrome_subsequence("aaabaaa")
```

```python
""" Given a string S and a string T, find the minimum window in S which will contain all
the characters in T in complexity O(n).

Example:

Input: S = "ADOBECODEBANC", T = "ABC"
Output: "BANC" 
Time Complexity: O(|S| + |T|)O(∣S∣+∣T∣) where |S| and |T| represent the lengths of strings
"""

def minWindow(s, t):
    if not t or not s:
        return ""
    # count of all unique chars in target string
    counter_dict = Counter(t)
    # total required characters from target string
    required = len(counter_dict)
    left = right = 0
    # did we found all unique characters from target string ?
    found = 0
    # count of unique characters in current window
    window_count = {}
    # final result
    result = float('inf'), 0, 0
    
    while right < len(s):
        curr_char = s[right]
        window_count[curr_char] = window_count.get(curr_char, 0) + 1
        
        if curr_char in counter_dict and window_count[curr_char] == counter_dict[curr_char]:
            found += 1
        
        while left<=right and found == required:
            # we found all unique chars, and try finding smaller window now.
            new_char = s[left]
            # store smallest result first
            if right - left + 1 < result[0]:
                result = (right-left+1, left, right)
            
            window_count[new_char] -= 1
            if new_char in counter_dict and window_count[new_char] < counter_dict[new_char]:
                found -= 1
            left += 1
        right += 1
    return "" if result[0] == float('inf') else s[result[1]:result[2]+1]

minWindow('ABAACBAB', 'ABC')
```

```python
""" 
Given 2 words, return true if second word has a substring that is also
an anagram of word 1.

LGE , GOOGLE- True
GEO, GOOGLE - False

TODO: similar to min-wondow problem
"""

def anagram(word, strs):
    if not word or not strs:
        return
    words_dict = Counter(word)
    required = len(words_dict)
    found = 0
    left = right = 0
    window_count = {}
    result = float('inf'), 0, 0
    
    while right < len(strs):
        curr_char = strs[right]
        window_count[curr_char] = window_count.get(curr_char, 0) + 1
        
        if curr_char in words_dict and window_count[curr_char] == words_dict[curr_char]:
            found += 1
        
        while left <= right and found == required:
            new_char = strs[left]
            if right - left + 1 < result[0]:
                result = (right-left+1, left, right)
            window_count[new_char] -= 1
            if new_char in words_dict and window_count[new_char] < words_dict[new_char]:
                found -= 1
            left += 1
        right += 1
    bucket = strs[result[1]: result[2]+1]
    if Counter(bucket) == words_dict:
        return True
    return False

# anagram('LGE', 'GOOGLE')
anagram('GEO', 'GOOGLE')
```

```python
"""
Given a non-empty array nums contains positive integers and a positive integer target.
Find the first subarray in nums that sums up to target and return the begin and
end index of this subarray. If there is no such subarray, return [-1, -1].

For example:

input: nums=[4, 3, 5, 7, 8], target = 12
return: [0, 2]
explain: 4 + 3 + 5 = 12. Although 5 + 7 = 12, [4, 3, 5] is the first subarray that sums up to 12.

input: nums[1, 2, 3, 4], target = 15
return: [-1, -1]
explain: Thers is no such subarray that sums up to 15.

"""

def find_target(nums, target):
    start = 0
    end = 1
    curr_total = nums[start] + nums[end]
    while end < len(nums):
        if curr_total == target:
            return [start, end]
        elif curr_total > target:
            curr_total -= nums[start]
            start += 1
        elif curr_total < target:
            end += 1
            if end < len(nums):
                curr_total += nums[end]
    return [-1, -1]

find_target([4, 3, 5, 7, 8], 12)
find_target([1, 2, 3, 4], 15)
```

```python
""" 841. Keys and Rooms

Input: [[1],[2],[3],[]]
Output: true

Input: [[1,3],[3,0,1],[2],[0]]
Output: false
"""
def canVisitAllRooms(rooms):
    seen = [False] * len(rooms)
    seen[0] = True
    stack = [0]
    #At the beginning, we have a todo list "stack" of keys to use.
    #'seen' represents at some point we have entered this room.
    while stack:
        node = stack.pop() # get the next key 'node'
        for nei in rooms[node]: # For every key in room # 'node'...
            if not seen[nei]: # ... that hasn't been used yet
                seen[nei] = True # mark that we've entered the room
                stack.append(nei) # add the key to the todo list
    return all(seen) # Return true iff we've visited every room
```

```python
""" 403.frog jump """
# hashmap
def canCross1(stones):
    if stones[1] != 1:
        return False
    d = {x: set() for x in stones}
    d[1].add(1)
    for x in stones[:-1]:
        for j in d[x]:
            for k in range(j-1, j+2):
                if k > 0 and x+k in d:
                    d[x+k].add(k)
    return bool(d[stones[-1]])

# stack
def canCross(stones):
    """
    :type stones: List[int]
    :rtype: bool
    """
    stone_set, fail = set(stones), set()
    stack = [(0, 0)]
    while stack:
        stone, jump = stack.pop()
        for j in (jump-1, jump, jump+1):
            s = stone + j
            if j > 0 and s in stone_set and (s, j) not in fail:
                if s == stones[-1]:
                    return True
                stack.append((s, j))
        fail.add((stone, jump))
    return False

stones = [0,1,3,5,6,8,12,17]
canCross(stones)
```

```python
""" 681. Next closest time """
def nextClosestTime(time):
    """
    :type time: str
    :rtype: str
    """
    hour, minute = time.split(":")
    print(">> hour: %s" %hour)
    print(">> min: %s" %minute)
    # Generate all possible 2 digit values
    # There are at most 16 sorted values here
    nums = sorted(set(hour + minute))
    two_digit_values = [a+b for a in nums for b in nums]
    print(two_digit_values)
    # Check if the next valid minute is within the hour
    i = two_digit_values.index(minute)
    if i + 1 < len(two_digit_values) and two_digit_values[i+1] < "60":
        return hour + ":" + two_digit_values[i+1]

    # Check if the next valid hour is within the day
    i = two_digit_values.index(hour)
    if i + 1 < len(two_digit_values) and two_digit_values[i+1] < "24":
        return two_digit_values[i+1] + ":" + two_digit_values[0]

    # Return the earliest time of the next day
    return two_digit_values[0] + ":" + two_digit_values[0]

time = "23:59"
nextClosestTime(time)
```

```python
""" minimum swaps to sort array """
def minSwaps(array):
    n = len(array) 
    arrpos = [(i, n) for i, n in enumerate(array)]
    pp.pprint(arrpos)
    arrpos.sort(key = lambda x:x[1]) 
    pp.pprint(arrpos)
    visited = {k:False for k in range(n)}
    ans = 0
    for i in range(n):
        if visited[i] or arrpos[i][0] == i: 
            continue
        cycle_size = 0
        j = i 
        while not visited[j]: 
            visited[j] = True
            j = arrpos[j][0] 
            cycle_size += 1

        if cycle_size > 0: 
            ans += (cycle_size - 1) 
    return ans

# Driver Code      
arr = [27, 21, 23, 22, 24, 25, 26, 28]
print(minSwaps(arr)) 
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.end = False

    def insert(self, word):
        curr = self
        for ch in word:
            if ch not in curr.children:
                curr.children[ch] = TrieNode()
            curr = curr.children[ch]
        curr.end = True
```

```python
""" Given two strings A and B, what's the longest prefix of A which is suffix of B """

def longest_prefix_suffix(strs1, strs2):
    root = TrieNode()
    root.insert(strs1)
    curr = root
    res = []
    for ch in strs2:
        if ch in curr.children:
            res.append(ch)
            curr = curr.children[ch]
        else:
            curr = root
            res = []
            if ch in curr.children:
                res.append(ch)
                curr = curr.children[ch]
    return ''.join(res) if res else 'Not Found'

print(longest_prefix_suffix('flower', 'flsrflow'))
print(longest_prefix_suffix('flower', 'werffl'))
print(longest_prefix_suffix('flower', 'werflowt'))
print(longest_prefix_suffix('flower', 'oflower'))
```

```python
""" longest common substring """
def longest_substring(strs1, strs2):
    m = len(strs1)
    n = len(strs2)
    dp = [[0] * (m+1) for _ in range(n+1)]
    #pp.pprint(dp)
    res = 0
    result = []
    for i in range(1, len(strs1)):
        for j in range(1, len(strs2)):
            if strs1[i-1] == strs2[j-1]:
                dp[i][j] = dp[i-1][j-1] +1;
                result.append(strs1[i-1])
                if res < dp[i][j]:
                    res = dp[i][j]
    return res, result

def longest_subsequence(strs1, strs2):
    m = len(strs1)
    n = len(strs2)
    table = [[0] * (n+1) for _ in range(m+1)]
    #pp.pprint(table)
    res = 0
    result = []
    for i in range(1, m+1):
        for j in range(1, n+1):
            if i == 0 or j == 0:
                table[i][j] = 0
            elif strs1[i-1] == strs2[j-1]:
                table[i][j] = table[i-1][j-1] + 1
            else:
                table[i][j] = max(table[i-1][j], table[i][j-1])
            if res < table[i][j]:
                res = table[i][j]
                result.append(strs1[i-1])
    return res, result
    
print(longest_substring('abcdaf', 'zbcdf'))
print(longest_subsequence('abcdaf', 'acbcf'))
print(longest_subsequence('ABCDGHLQR', 'AEDPHR'))
```

```python
"""
Given an array of unique positive integers, for each integer,
return the first larger element in the array to the left. If there's no larger element, return -1 for it.
Example:
arr = [70, 60, 40, 50, 35, 90, 10]
ans = [-1, 70, 60, 60, 50, -1, 90]

"""

def left_largest(nums):
    result = [-1]
    stack = [nums[0]]
    for n in nums[1:]:
        if n > stack[-1]:
            while stack and n > stack[-1]:
                tmp = stack.pop()
        if not stack:
            result.append(-1)
        else:
            result.append(stack[-1])
        stack.append(n)
    return result

nums = [70, 60, 40, 50, 35, 90, 10]
left_largest(nums)
```

```python
""" KMP pattern search """

def kmp_pattern(text, pattern):
    
    def compute_array(pattern):
        lps = [0] * len(pattern)
        ind = 0
        i = 1
        while i < len(pattern):
            if pattern[i] == pattern[ind]:
                lps[i] = ind + 1
                i += 1
                ind += 1
            else:
                if ind != 0:
                    ind = lps[ind-1]
                else:
                    lps[i] = 0
                    i += 1
        return lps

    lps = compute_array(pattern)
    pp.pprint(lps)    
    i = 0
    j = 0
    while i<len(text) and j<len(pattern):
        if text[i] == pattern[j]:
            i += 1
            j += 1
        else:
            if j != 0:
                j = lps[j-1]
            else:
                i += 1
    if j == len(pattern):
        return True
    return False

text = 'abcxabcdabcdabcy'
pattern = 'abcdabcy'
kmp_pattern(text, pattern)
```

```python
""" check if number is power of 7 """

def power_of(x, y=7):
    if x == 0:
        return True
    while (x%y == 0):
        x = x//y
    return x==1

print(power_of(49))
print(power_of(23))
print(power_of(0))
```
